```python
csv_path = '/storage/yue/data/CBISDDSM/ddsm_csv_all.csv'
df = pd.read_csv(csv_path).reset_index(drop=True)
print(df.columns.tolist())
df = df[['patient_id', 'image file path', 'abnormality type']].drop_duplicates().reset_index(drop=True)
print(set(df['abnormality type']))
index_list = df[df['abnormality type']=='calcification'].index.tolist()
df.loc[index_list, 'label']=0
print(len(index_list))
index_list = df[df['abnormality type']=='mass'].index.tolist()
df.loc[index_list, 'label']=1
print(len(index_list))
df
```

```python
num_patients = len(list(set(df['patient_id'].tolist())))
print(num_patients)
num_patients_train = int(num_patients*0.8)
num_patients_val = num_patients-num_patients_train
print(num_patients_train, num_patients_val)
```

```python
all_patients = list(set(df['patient_id'].tolist()))
random.shuffle(all_patients)
train_patients = all_patients[:num_patients_train]
val_patients = all_patients[num_patients_train:]
print(len(train_patients))
print(len(val_patients))
```

```python
index_list = df[df['patient_id'].isin(train_patients)].index.tolist()
df.loc[index_list, 'split']='train'
print(len(index_list))
index_list = df[df['patient_id'].isin(val_patients)].index.tolist()
df.loc[index_list, 'split']='val'
print(len(index_list))
df
```

```python
print(df[(df['split']=='train') & (df['label']==0)].shape[0], df[(df['split']=='train') & (df['label']==1)].shape[0])
print(df[(df['split']=='val') & (df['label']==0)].shape[0], df[(df['split']=='val') & (df['label']==1)].shape[0])
```
